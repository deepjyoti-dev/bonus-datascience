**Binary Image**

1 bit Monochrome Image ( Bitmap Image )

Assume image is a grid of 8 x 8 pixels

8*8 = 64 pixels in total

Each pixels require 1 bit, 

to store either BLACK( True = 1 ) or WHITE ( False = 0 )


**Grayscale Image**

8 bit Grayscale Image

Assume image is a grid of 8 x 8 pixels

8*8 = 64 pixels in total

Each pixels require 0 to 255 ( Unsigned Integer )

to store either BLACK(0) or WHITE (255)Or GRAY(128)

 

**Color Image**

8 bit Color Image

Assume image is a grid of 8 x 8 pixel

8*8 = 64 pixels in total

# Image Processing using Scikit-image and Numpy Array



Color-image data for single image is typically stored in three dimensions. 

Each image is a three-dimensional array of (height, width, channels), 
where the channels are usually red, green, and blue (RGB) values. 

One 256x128 ( width, height ) RGB images would have shape (128, 256, 3) ( height, width, channels ). 

(An extended representation is RGBA, where the Aâ€“alphaâ€“denotes the level of opacity.)

One 256x128 ARGB images would have shape (128, 256, 4). 


Color-image data for multiple images is typically stored in four dimensions. 

A collection of images is then just (image_number, height, width, channels). 

One thousand 256x128 RGB images would have shape (1000, 128, 256, 3).

# Reading the stars from the sky

```python
from matplotlib import pyplot as plt
from skimage import data
from skimage.feature import blob_dog, blob_log, blob_doh
from skimage.color import rgb2gray
from skimage.io import imread
import math
```

```python
img = imread('/content/drive/My Drive/day0/wint_sky.gif','gray')
plt.imshow(img, "gray")
plt.show()
```

```python
print(type(img))
print(img.shape)
```

```python
# Blobs are bright on dark or dark on bright regions in an image.
# 
blobs_log = blob_log(img, max_sigma=30, num_sigma=10, threshold=.1)
numrows = len(blobs_log)
print("Numer of stars counted : " ,numrows)
```

```python
fig, ax = plt.subplots(1, 1)
plt.imshow(img, 'gray')
for blob in blobs_log:
    y, x, r = blob
    c = plt.Circle((x, y), r+5, color='lime', linewidth=2, fill=False)
    ax.add_patch(c)

```
